Copying the dataframe into test

```python
test=df.copy()
```

Label encoding columns where the feature type is object.

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
#Declaring the columns where the features are of object/string type
cols = ['director_name','actor_2_name','genres','language','country','content_rating','movie_category','actor_1_name','plot_keywords','actor_3_name','movie_title']
test[cols] = test[cols].apply(label_encoder.fit_transform)
```

```python
test.head()
```

Dropping the Unnamed column from the test dataframe. This means this feature will be discarded

```python
#Dropping the column
test=test.drop(['Unnamed: 0'], axis=1)
```

CHecking the information about the test dataframe after dropping the column and label encoding

```python
test.info()
```

```python
test.head()
```

Data Normalization: Here all the features (exept genres) are normalized using the MinMaxScalar library. After normalization the values should be between 0 and 1

```python

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
#performing normalization
scaled = scaler.fit(test.drop('genres',axis=1)).transform(test.drop('genres',axis=1))
```

Preparaing the new dataframe with normalized values and Genre column separated.

```python
final_version=test.copy()
final_version=final_version.drop('genres',axis=1)
df_scaled = pd.DataFrame(scaled, columns=final_version.columns)
df_scaled.head()
```

From this observation we can see that all the features now have the same data type

```python
df_scaled.info()
```

Assigning values to X and y. y has the genre predictions

```python
#df_scaled.drop('director_facebook_likes',axis=1)
x=df_scaled
#y is assigned the genre column of test dataframe which is not normalized but label encoded
y=test['genres']
print(y)
```

Dataset splitting: Here splitting takes place such that 80% of the original data is contained by the training dataset.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

SVM model is trained using rbf kernel

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import time
#constructing the model with the rbf kernel
svc=SVC(kernel="rbf",probability=False) 
svc.fit(x_train, y_train)
#using time for inference
start_time = time.time()
y_pred=svc.predict(x_test)
inference_time = time.time() - start_time
print("Inference time: {:.5f} seconds".format(inference_time))
print('Model accuracy : {0:0.3f}'. format(accuracy_score(y_test, y_pred)))
```

Getting the parameters of the model

```python
svc.get_params()
```

Training and evaluating the model with linear kernel 

```python
svc=SVC(kernel="linear", probability=True) 
svc.fit(x_train, y_train)
start_time = time.time()
y_pred=svc.predict(x_test)
inference_time = time.time() - start_time
print("Inference time: {:.5f} seconds".format(inference_time))
#evaluation
print('Model accuracy : {0:0.3f}'. format(accuracy_score(y_test, y_pred)))
```

```python
svc.get_params()
```

Training and evaluating the model with poly kernel 

```python
svc=SVC(kernel="poly",probability=False) 
svc.fit(x_train, y_train)
start_time = time.time()
y_pred=svc.predict(x_test)
inference_time = time.time() - start_time
print("Inference time: {:.5f} seconds".format(inference_time))
#evaluation
print('Model accuracy : {0:0.3f}'. format(accuracy_score(y_test, y_pred)))
```
